```python
def resamples(samples):
    x = np.vstack((samples['pos_samples']['X'],samples['neg_samples']['X']))
    y = np.vstack((samples['pos_samples']['X'],samples['neg_samples']['X']))
    print('采样前正负样本比例：正：负 = ',samples['pos_samples']['X'].shape[0],':',samples["neg_samples"]["X"].shape[0])
    rus = RandomUnderSampler(random_state=42, ratio={0.0 : int(1.1 * samples['pos_samples']['X'].shape[0])})
    x_res,y_res = rus.fit_sample(x,y)
    return x_res,y_res
```
